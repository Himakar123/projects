```python

def conv(Image,filter):
   Image= np.array(Image)
   n=Image.shape[0]
   m=Image.shape[1]
   f=filter.shape[0]
   new_image =np.zeros((n-f+1,m-f+11))
   for i in range(n-f+1):
       for j in range(m-f+1):
           arr1=np.array(Image[i:i+f, j:j+f])
           ans= np.sum(arr1*filter)
           new_image[i,j]=ans
   return new_image
```
